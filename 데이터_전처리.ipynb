```python
import os
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Specify the path to the folder in your Drive
folder_path = '/content/drive/MyDrive/ML_dataset' # Assuming 'ML_dataset' is directly in My Drive

# List all files in the directory
try:
    file_list = os.listdir(folder_path)
    print(f"Files in {folder_path}:", file_list)

    # Load each file into a pandas DataFrame and save as CSV
    dataframes = {}
    for file in file_list:
        file_path = os.path.join(folder_path, file)
        df_name = os.path.splitext(file)[0]  # Use filename without extension as DataFrame name

        try:
            if file.endswith('.csv'):
                dataframes[df_name] = pd.read_csv(file_path)
                print(f"\nDataFrame '{df_name}' loaded from CSV:")
                display(dataframes[df_name].head())
            elif file.endswith('.json'):
                dataframes[df_name] = pd.read_json(file_path)
                csv_file_path = os.path.join(folder_path, f"{df_name}.csv")
                dataframes[df_name].to_csv(csv_file_path, index=False)
                print(f"\nDataFrame '{df_name}' loaded from JSON and saved as CSV:")
                display(dataframes[df_name].head())
            elif file.endswith('.xlsx'):
                dataframes[df_name] = pd.read_excel(file_path)
                csv_file_path = os.path.join(folder_path, f"{df_name}.csv")
                dataframes[df_name].to_csv(csv_file_path, index=False)
                print(f"\nDataFrame '{df_name}' loaded from Excel and saved as CSV:")
                display(dataframes[df_name].head())

        except Exception as e:
            print(f"Error loading {file}: {e}")

except FileNotFoundError:
    print(f"Error: The folder '{folder_path}' was not found in your Google Drive.")
except Exception as e:
    print(f"An error occurred: {e}")
```

# d01 전처리





```python
# Specify the path to the d01_reframing_dataset.csv file
file_path_d01 = os.path.join(folder_path, 'd01_reframing_dataset.csv')

# Load the d01_reframing_dataset.csv file into a pandas DataFrame
try:
    df_d01 = pd.read_csv(file_path_d01)
    print("DataFrame 'd01_reframing_dataset' loaded from CSV:")
    display(df_d01.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path_d01}' was not found.")
except Exception as e:
    print(f"Error loading d01_reframing_dataset.csv: {e}")

# Rename the column 'thinking_traps_addressed' to 'distortion_type'
df_d01 = df_d01.rename(columns={'thinking_traps_addressed': 'distortion_type'})

# Display the head of the dataframe with the renamed column
print("DataFrame with renamed column:")
display(df_d01.head())

# Update the dataframe in the dictionary
dataframes['d01_reframing_dataset'] = df_d01
```

```python
# Split the 'distortion_type' column by comma and create a list of distortion types for each row
distortion_lists = dataframes['d01_reframing_dataset']['distortion_type'].str.split(',').tolist()

# Get all unique distortion types
all_distortions = [item for sublist in distortion_lists for item in sublist]
unique_distortions = sorted(list(set(all_distortions)))

# Create new columns for each unique distortion type and initialize with 0
for distortion in unique_distortions:
    dataframes['d01_reframing_dataset'][distortion.strip()] = 0

# Iterate through each row and mark the presence of distortion types with 1
for index, row in dataframes['d01_reframing_dataset'].iterrows():
    distortions_in_row = [d.strip() for d in row['distortion_type'].split(',')]
    for distortion in distortions_in_row:
        if distortion in unique_distortions:
            dataframes['d01_reframing_dataset'].loc[index, distortion] = 1

# Create the 'has_distortion' column
dataframes['d01_reframing_dataset']['has_distortion'] = (dataframes['d01_reframing_dataset']['none'] == 0).astype(int)

# Drop the original 'distortion_type' and 'none' columns
dataframes['d01_reframing_dataset'] = dataframes['d01_reframing_dataset'].drop(columns=['distortion_type', 'none'])

# Display the head of the dataframe with the new columns
print("DataFrame with distortion type columns:")
display(dataframes['d01_reframing_dataset'].head())
```

```python
# Get the list of columns in the dataframe
all_columns = dataframes['d01_reframing_dataset'].columns.tolist()

# Remove the original columns and the newly added 'has_distortion' column
distortion_columns = [col for col in all_columns if col not in ['situation', 'thought', 'reframe', 'has_distortion']]

# Print the distortion type column names
print("Distortion type feature names:")
print(distortion_columns)
```

```python
# Get the dataframe with existing one-hot encoded distortion columns
df_d01 = dataframes['d01_reframing_dataset'].copy()

# Get the list of existing distortion columns (excluding original columns and 'has_distortion')
all_columns = df_d01.columns.tolist()
distortion_columns = [col for col in all_columns if col not in ['situation', 'thought', 'reframe', 'has_distortion']]

# Add 'mental filter' column if not already present
if 'mental filter' not in distortion_columns:
    df_d01['mental filter'] = 0
    distortion_columns.append('mental filter')

# Consolidate 'magnification' and 'catastrophizing' into 'magnification'
if 'catastrophizing' in df_d01.columns:
    if 'magnification' in df_d01.columns:
        df_d01['magnification'] = df_d01[['magnification', 'catastrophizing']].max(axis=1)
    else:
        df_d01['magnification'] = df_d01['catastrophizing']
    df_d01 = df_d01.drop(columns=['catastrophizing'])
    if 'catastrophizing' in distortion_columns:
        distortion_columns.remove('catastrophizing')
    if 'magnification' not in distortion_columns:
        distortion_columns.append('magnification')


# Consolidate 'personalizing' and 'blaming' into 'personalization and blaming'
columns_to_consolidate_pb = [col for col in ['personalizing', 'blaming'] if col in df_d01.columns]
if columns_to_consolidate_pb:
    df_d01['personalization and blaming'] = df_d01[columns_to_consolidate_pb].max(axis=1)
    df_d01 = df_d01.drop(columns=columns_to_consolidate_pb)
    for col in columns_to_consolidate_pb:
        if col in distortion_columns:
            distortion_columns.remove(col)
    if 'personalization and blaming' not in distortion_columns:
        distortion_columns.append('personalization and blaming')


# Ensure 'has_distortion' is correctly calculated based on the presence of any distortion
# This assumes 'none' column was dropped in a previous step, so we sum up the one-hot encoded columns
distortion_cols_after_consolidation = [col for col in df_d01.columns if col not in ['situation', 'thought', 'reframe']] # Exclude original and has_distortion
if 'has_distortion' in distortion_cols_after_consolidation:
    distortion_cols_after_consolidation.remove('has_distortion')

if distortion_cols_after_consolidation:
    df_d01['has_distortion'] = (df_d01[distortion_cols_after_consolidation].sum(axis=1) > 0).astype(int)
else:
     df_d01['has_distortion'] = 0 # If no distortion columns, assume no distortion

# Reorder columns to move 'has_distortion' after 'reframe'
cols = df_d01.columns.tolist()
if 'has_distortion' in cols:
    has_distortion_col = cols.pop(cols.index('has_distortion'))
    try:
        reframe_index = cols.index('reframe')
        cols.insert(reframe_index + 1, has_distortion_col)
    except ValueError:
        # If 'reframe' is not found, just append 'has_distortion' at the end
        cols.append(has_distortion_col)
    df_d01 = df_d01[cols]


# Update the dataframe in the dictionary with the new name
dataframes['d01_preprocessed'] = df_d01

# Display the head of the dataframe with the new columns
print("DataFrame 'd01_preprocessed' with consolidated distortion type columns:")
display(dataframes['d01_preprocessed'].head())
```

```python
# Get the preprocessed dataframe
df_d01_processed = dataframes['d01_preprocessed'].copy()

# Remove rows where 'negative feeling or emotion' is 1
df_d01_processed = df_d01_processed[df_d01_processed['negative feeling or emotion'] == 0]

# Drop the 'negative feeling or emotion' column
df_d01_processed = df_d01_processed.drop(columns=['negative feeling or emotion'])

# Update the dataframe in the dictionary
dataframes['d01_preprocessed'] = df_d01_processed

# Display the head of the modified dataframe
print("DataFrame 'd01_preprocessed' after removing rows and column 'negative feeling or emotion':")
display(dataframes['d01_preprocessed'].head())
```

```python
# Get the d01 dataframe
df_d01 = dataframes['d01_reframing_dataset'].copy()

# Get unique situations in the order they appear
unique_situations_ordered = df_d01['situation'].unique()

# Create a mapping from situation to case_id
situation_to_id = {situation: i for i, situation in enumerate(unique_situations_ordered)}

# Create the 'case_id' column using the mapping
df_d01['case_id'] = df_d01['situation'].map(situation_to_id)

# Get the list of columns
cols = df_d01.columns.tolist()

# Move 'case_id' to the front of the list
cols = ['case_id'] + [col for col in cols if col != 'case_id']

# Reindex the dataframe with the new column order
df_d01 = df_d01[cols]

# Update the dataframe in the dictionary
dataframes['d01_reframing_dataset'] = df_d01

# Display the head of the dataframe with the new 'case_id' column
print("DataFrame 'd01_reframing_dataset' with 'case_id' column:")
display(dataframes['d01_reframing_dataset'].head())
```

# d02 전처리

```python
# Merge the two dataframes based on 'Id_Number'
merged_df = pd.merge(dataframes['d02_Annotated_data'], dataframes['d02_Therapist_responses'], on='Id_Number', how='inner')

# Sort by 'Id_Number' in ascending order
merged_df = merged_df.sort_values(by='Id_Number')

# Drop the duplicate 'Question' column (assuming 'Patient Question' is the one to keep)
merged_df = merged_df.drop(columns=['Question'])

# Display the head of the merged dataframe
print("Merged DataFrame:")
display(merged_df.head())
```

```python
# Define the path to save the merged dataframe
save_path = os.path.join(folder_path, 'd02_merged_data.csv')

# Save the merged dataframe to a CSV file
merged_df.to_csv(save_path, index=False)

print(f"Merged dataframe saved to: {save_path}")
```

```python
# Get all unique distortion types from both columns
all_distortions_d02 = pd.concat([merged_df['Dominant Distortion'], merged_df['Secondary Distortion (Optional)']]).dropna().unique()

# Create new columns for each unique distortion type and initialize with 0
for distortion in all_distortions_d02:
    merged_df[distortion] = 0

# Iterate through each row and mark the presence of distortion types with 1
for index, row in merged_df.iterrows():
    dominant_distortion = row['Dominant Distortion']
    secondary_distortion = row['Secondary Distortion (Optional)']

    if pd.notna(dominant_distortion):
        merged_df.loc[index, dominant_distortion] = 1
    if pd.notna(secondary_distortion):
        merged_df.loc[index, secondary_distortion] = 1

# Drop the original distortion columns
merged_df = merged_df.drop(columns=['Dominant Distortion', 'Secondary Distortion (Optional)'])

# Rename the 'Answer' column to 'reframe'
merged_df = merged_df.rename(columns={'Answer': 'reframe'})

# Create the 'has_distortion' column based on 'No Distortion'
merged_df['has_distortion'] = (merged_df['No Distortion'] == 0).astype(int)

# Drop the 'No Distortion' column
merged_df = merged_df.drop(columns=['No Distortion'])


# Display the head of the modified dataframe
print("DataFrame with one-hot encoded distortions and renamed column:")
display(merged_df.head())
```

```python
# Rename the 'Patient Question' column to 'situation'
merged_df = merged_df.rename(columns={'Patient Question': 'situation'})

# Rename the 'Distorted part' column to 'thought'
merged_df = merged_df.rename(columns={'Distorted part': 'thought'})

# Display the head of the modified dataframe
print("DataFrame with renamed columns:")
display(merged_df.head())
```

```python
# Get the list of all columns in the merged_df
all_merged_columns = merged_df.columns.tolist()

# Remove columns that are not distortion types
distortion_columns_merged = [col for col in all_merged_columns if col not in ['Id_Number', 'Patient Question', 'Distorted part', 'reframe', 'has_distortion']]

# Print the distortion type column names
print("Distortion type feature names in merged_df:")
print(distortion_columns_merged)
```

```python
# Convert all column names to lowercase
merged_df.columns = merged_df.columns.str.lower()

# Rename 'fortune-telling' to 'fortune telling'
merged_df = merged_df.rename(columns={'fortune-telling': 'fortune telling'})

# Create a new column 'disqualifiying the positive' and set all values to 0
merged_df['disqualifiying the positive'] = 0

# Rename 'id_number' to 'case_id'
merged_df = merged_df.rename(columns={'id_number': 'case_id'})

# Update the dataframe in the dictionary with the new name
dataframes['d02_preprocessed'] = merged_df

# Display the head of the modified dataframe
print("Merged DataFrame after renaming and adding column:")
display(dataframes['d02_preprocessed'].head())
```

```python
# Get the d02_preprocessed dataframe
df_d02_processed = dataframes['d02_preprocessed']

# Get the case_id column and sort it
case_ids = df_d02_processed['case_id'].sort_values().tolist()

# Check the range of case_ids
min_id = case_ids[0] if case_ids else None
max_id = case_ids[-1] if case_ids else None

print(f"Case ID ranges from {min_id} to {max_id}.")

# Check for missing numbers
if case_ids:
    expected_ids = list(range(min_id, max_id + 1))
    missing_ids = sorted(list(set(expected_ids) - set(case_ids)))

    if missing_ids:
        print(f"Missing Case IDs: {missing_ids}")
    else:
        print("No missing Case IDs found.")
else:
    print("No case IDs found in the dataframe.")
```

```python
# Get the d02_preprocessed dataframe
df_d02_processed = dataframes['d02_preprocessed'].copy()

# Reorder columns to move 'has_distortion' after 'reframe'
cols = df_d02_processed.columns.tolist()
if 'has_distortion' in cols:
    cols.remove('has_distortion')
    try:
        reframe_index = cols.index('reframe')
        cols.insert(reframe_index + 1, 'has_distortion')
    except ValueError:
        # If 'reframe' is not found, just append 'has_distortion' at the end
        cols.append('has_distortion')

# Reindex the dataframe with the new column order
df_d02_processed = df_d02_processed[cols]

# Update the dataframe in the dictionary
dataframes['d02_preprocessed'] = df_d02_processed

# Display the head of the modified dataframe
print("DataFrame 'd02_preprocessed' after reordering:")
display(dataframes['d02_preprocessed'].head())
```

```python
# Specify the path to the d03_cbt_df.csv file
file_path_d03 = os.path.join(folder_path, 'd03_cbt_df.csv')

# Load the d03_cbt_df.csv file into the dataframes dictionary
try:
    dataframes['d03_cbt_df'] = pd.read_csv(file_path_d03)
    print("DataFrame 'd03_cbt_df' reloaded from CSV:")
    display(dataframes['d03_cbt_df'].head())
except FileNotFoundError:
    print(f"Error: The file '{file_path_d03}' was not found.")
except Exception as e:
    print(f"Error reloading d03_cbt_df.csv: {e}")
```

```python
# Get unique situations in the order they appear
unique_situations_ordered = dataframes['d03_cbt_df']['situation'].unique()

# Create a mapping from situation to case_id
situation_to_id = {situation: i for i, situation in enumerate(unique_situations_ordered)}

# Create the 'case_id' column using the mapping
dataframes['d03_cbt_df']['case_id'] = dataframes['d03_cbt_df']['situation'].map(situation_to_id)

# Get the list of columns
cols = dataframes['d03_cbt_df'].columns.tolist()

# Move 'case_id' to the front of the list
cols = ['case_id'] + [col for col in cols if col != 'case_id']

# Reindex the dataframe with the new column order
dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'][cols]

# Display the head of the dataframe with the new 'case_id' column
print("DataFrame with 'case_id' column:")
display(dataframes['d03_cbt_df'].head())
```

```python
# Drop columns
dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'].drop(columns=['domain', 'prefix', 'input_text'])

# Rename columns
dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'].rename(columns={'negative_thought': 'thought', 'target_text': 'reframe'})

# Split the 'distortions' column by comma and create a list of distortion types for each row
distortion_lists_d03 = dataframes['d03_cbt_df']['distortions'].str.split(',').tolist()

# Get all unique distortion types
all_distortions_d03 = [item for sublist in distortion_lists_d03 for item in sublist]
unique_distortions_d03 = sorted(list(set(all_distortions_d03)))

# Create new columns for each unique distortion type and initialize with 0
for distortion in unique_distortions_d03:
    dataframes['d03_cbt_df'][distortion.strip()] = 0

# Iterate through each row and mark the presence of distortion types with 1
for index, row in dataframes['d03_cbt_df'].iterrows():
    distortions_in_row = [d.strip() for d in row['distortions'].split(',')]
    for distortion in distortions_in_row:
        if distortion in unique_distortions_d03:
            dataframes['d03_cbt_df'].loc[index, distortion] = 1

# Identify the unnamed column (assuming it's the one with an empty string name)
# Check if an unnamed column exists before trying to access it
unnamed_columns = [col for col in dataframes['d03_cbt_df'].columns if col.strip() == '']
if unnamed_columns:
    unnamed_column = unnamed_columns[0]
    # Create the 'has_distortion' column based on the unnamed column
    dataframes['d03_cbt_df']['has_distortion'] = (dataframes['d03_cbt_df'][unnamed_column] == 0).astype(int)
    # Drop the original 'distortions' column and the unnamed column
    dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'].drop(columns=['distortions', unnamed_column])
else:
    # If no unnamed column, create 'has_distortion' assuming no 'none' equivalent
    dataframes['d03_cbt_df']['has_distortion'] = 1 # Assuming if no unnamed column, all have distortion
    # Drop the original 'distortions' column
    dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'].drop(columns=['distortions'])


# Get current columns
cols = dataframes['d03_cbt_df'].columns.tolist()

# Move 'has_distortion' after 'reframe'
if 'has_distortion' in cols:
    cols.remove('has_distortion')
    try:
        reframe_index = cols.index('reframe')
        cols.insert(reframe_index + 1, 'has_distortion')
    except ValueError:
        # If 'reframe' is not found, just append 'has_distortion' at the end
        cols.append('has_distortion')


# Reindex the dataframe with the new column order
dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'][cols]

# Display the head of the modified dataframe
print("DataFrame d03_cbt_df after cleaning, one-hot encoding, and reordering:")
display(dataframes['d03_cbt_df'].head())

# Get and print distortion type column names
all_columns_d03 = dataframes['d03_cbt_df'].columns.tolist()
distortion_columns_d03 = [col for col in all_columns_d03 if col not in ['case_id', 'situation', 'emotions', 'thought', 'reframe', 'has_distortion']]

print("\nDistortion type feature names in d03_cbt_df:")
print(distortion_columns_d03)
```

```python
# Convert all column names to lowercase
dataframes['d03_cbt_df'].columns = dataframes['d03_cbt_df'].columns.str.lower()

# Rename columns
dataframes['d03_cbt_df'] = dataframes['d03_cbt_df'].rename(columns={
    'jumping to conclusions (mind reading)': 'mind reading',
    'jumping to conclusions (fortune teller error)': 'fortune telling'
})

# Display the head of the modified dataframe
print("DataFrame d03_cbt_df after renaming columns:")
display(dataframes['d03_cbt_df'].head())
```
